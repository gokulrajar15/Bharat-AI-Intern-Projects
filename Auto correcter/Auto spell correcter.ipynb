# get close matches function used in this problem

```python
from difflib import get_close_matches

def autocorrect(word):
    # If the word is already in the word list, return it
    if word in word_list:
        return word
    # Otherwise, try to find a close match using get_close_matches
    close_matches = get_close_matches(word, word_list, n=2, cutoff=0.8)
    if len(close_matches) > 0:
        return close_matches[0]
    # If no close match is found, return the original word
    return word
```
